```python
# using where


df["is_LegalAffairs"] = np.where(df["Job Category"] == "Legal Affairs", 1, 0)
```

```python
# using mask

df['Civil Service Title'].mask(df['Civil Service Title'] == 'SENIOR IT ARCHITECT', 0, inplace=False)
```
